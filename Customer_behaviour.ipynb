```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x : x.fillna(x.median()))
```

```python
import re

# 1. Strip leading/trailing spaces
df.columns = df.columns.str.strip()

# 2. Replace ALL types of spaces (normal, multiple, unicode) with a single underscore
df.columns = df.columns.str.replace(r'\s+', '_', regex=True)

# 3. Remove any characters that are not letters/numbers/underscores
df.columns = df.columns.str.replace(r'[^0-9a-zA-Z_]', '', regex=True)

# 4. Convert to lowercase
df.columns = df.columns.str.lower()

df.columns
```

```python
import re

clean_cols = []

for col in df.columns:
    # 1. Remove ALL whitespace characters (space, tabs, unicode, invisible)
    col = re.sub(r'\s+', '', col)
    
    # 2. Remove surrounding underscores caused by cleanup
    col = col.strip('_')
    
    # 3. Convert to lowercase
    col = col.lower()
    
    clean_cols.append(col)

df.columns = clean_cols

df.columns
```

```python
df.columns = [
    'customer_id',
    'age',
    'gender',
    'item_purchased',
    'category',
    'purchase_amount',
    'location',
    'size',
    'color',
    'season',
    'review_rating',
    'subscription_status',
    'shipping_type',
    'discount_applied',
    'promo_code_used',
    'previous_purchases',
    'payment_method',
    'frequency_of_purchases'
]

df.columns
```

```python
#create a column age group
labels = ['Young adult', 'Adult', 'Middle-aged', 'Senior']
df['age_group'] = pd.qcut(df['age'], q=4, labels=labels)
```

```python
df[['age', 'age_group']].head(10)
```

```python
# create column purchase_frequency_days

frequency_mapping = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-Weekly': 14,
    'Annually': 365,
    'Every 3 Months': 90
}

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```
